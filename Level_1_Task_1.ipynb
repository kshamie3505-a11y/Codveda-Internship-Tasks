```python
soup.prettify
print(soup.prettify) # to display the code in a nested structure
```

**Finding other column headers from the html document**

```python
for book in soup.find_all("article", class_ = "product_pod"):
  title = book.h3.a.text # gives the book title
  price = book.find("p", class_="price_color").text
  rating = book.p['class'][1]
  availability = book.find("p",class_="instock availability").text.strip()
```

```python
books.append({
    "title" : title,
    "price" : price,
    "rating" : rating,
    "availability" : availability
})
```

**Converting html into a pandas DataFrame and storing it as a csv file**

```python
df = pd.DataFrame(books)
df.to_csv("books.csv", index= False_)
```
