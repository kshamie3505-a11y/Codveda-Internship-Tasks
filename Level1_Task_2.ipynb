**No missing values in the data from the above summary. In the case of any mssing values, use either drop or imputation method to fill them as shown below: # The error message is generated because the data has no missing values**

```python
# Drop columns with too many missing values # option 1
df = df.dropna(how='all')

# Filling in missing values(imputation) # option 2
df = df.fillna(df.mean()) # Using the mean or median for numerical rows/columns
df = df.fillna(df.mode().iloc[0]) # Using the mode for categorical rows/columns

```

**Dectecting outliers to show only columns with outliers**



```python
import numpy as np # Numerical library for python

# Select only numerical columns for outlier detection
numerical_cols = df.select_dtypes(include=np.number).columns # Identifies all columns with numerical data as outliers are typically found in numbers

outliers_dict = {} # Acts as a container for the identified outliers
for col in numerical_cols: # Looping through each numerical column to find outliers
    Q1 = df[col].quantile(0.25) # Calculates the first quartile
    Q3 = df[col].quantile(0.75) # Calculates the third quartile
    IQR = Q3 - Q1 # Represents middle 50% of the data

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col] # Acts as a filter, any value below or above is an outlier
    if not outliers.empty:
        outliers_dict[col] = outliers

if outliers_dict:
    print("Outliers detected in the following columns (showing first 5 for each if many):")
    for col, values in outliers_dict.items():
        print(f"Column '{col}':\n{values.head()}\n")
else:
    print("No significant outliers detected using the IQR method (1.5*IQR).")
```

**Detecting and removing the outliers at the same time.**

```python
import numpy as np

# Make a copy to avoid SettingWithCopyWarning
df_cleaned = df.copy() # Making a copy of the original df to preserve the original data

numerical_cols = df_cleaned.select_dtypes(include=np.number).columns # Outliers to be removed from numerical columns

# Iterate through numerical columns and filter out outliers
for col in numerical_cols:
    Q1 = df_cleaned[col].quantile(0.25) # Calculates the first quartile range
    Q3 = df_cleaned[col].quantile(0.75) # Calculates the third quartile range
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Keep only the rows where the column's value is within the bounds # Values below or above the bounds are filtered out as outliers
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]

print(f"Original DataFrame shape: {df.shape}") # Prints original columns before outlier removal
print(f"DataFrame shape after outlier removal: {df_cleaned.shape}") # Prints cleaned columns

# Display the first few rows of the cleaned DataFrame
display(df_cleaned.head())
```

**Convert categorical variables into numerical format using one-hot encoding or label encoding. Normalize or standardize numerical data.**

```python
# Applying the StandardScaler to numerical columns to standardize them

from sklearn.preprocessing import StandardScaler
import numpy as np

# Identify numerical columns in the cleaned DataFrame
numerical_cols_cleaned = df_cleaned.select_dtypes(include=np.number).columns # Identifies all the columns that contain numbers for the standardization

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply StandardScaler to the numerical columns
df_cleaned[numerical_cols_cleaned] = scaler.fit_transform(df_cleaned[numerical_cols_cleaned])

print("DataFrame after applying StandardScaler to numerical columns:")
display(df_cleaned.head())
```

```python
# Using One-hot encoding to convert categorical variables into numerical format
# Step 1 converting into False and True

# Identify categorical columns that are actually present in df_cleaned
categorical_cols_to_encode = ['State', 'International plan', 'Voice mail plan']
existing_categorical_cols = [col for col in categorical_cols_to_encode if col in df_cleaned.columns and df_cleaned[col].dtype == 'object']

# Convert existing categorical variables into numerical format using one-hot encoding
if existing_categorical_cols:
    df_cleaned = pd.get_dummies(df_cleaned, columns=existing_categorical_cols, drop_first=True) # drop_first=True to avoid multicollinearity
    print(f"Successfully encoded columns: {existing_categorical_cols}")
else:
    print("No specified categorical columns found or they are not of object dtype in the DataFrame.")

df_cleaned
```

```python
# Step 2 Converting the False and True into binary codes of 0 and 1
df_cleaned[['International plan_Yes', 'Voice mail plan_Yes','State_TX','State_UT','State_VA','State_VT','State_WA','State_WI','State_WV','State_WY']] = df_cleaned[['International plan_Yes', 'Voice mail plan_Yes','State_TX','State_UT','State_VA','State_VT','State_WA','State_WI','State_WV','State_WY']].astype(int)
df_cleaned
```

**Summary of Data cleaning and Preprocessing**

The following steps were performed to clean and preprocess the dataset:

1.  **Missing Data Handling**: Checked for missing values across all columns. It was confirmed that there were no missing values in the dataset, therefore no imputation or removal of rows/columns due to missing data was necessary.

2.  **Outlier Detection and Removal**: Outliers in numerical columns were identified and removed using the Interquartile Range (IQR) method. Any data points falling below `Q1 - 1.5 * IQR` or above `Q3 + 1.5 * IQR` were considered outliers and excluded from the dataset. This significantly reduced the number of rows from the original DataFrame.

3.  **Numerical Data Standardization**: The `StandardScaler` from `sklearn.preprocessing` was applied to all numerical columns. This transformed the data to have a mean of 0 and a standard deviation of 1, which is crucial for many machine learning algorithms.

4.  **Categorical Data Encoding**: Categorical variables (`State`, `International plan`, `Voice mail plan`) were converted into a numerical format using one-hot encoding. The `drop_first=True` argument was used to prevent multicollinearity. The resulting boolean columns from one-hot encoding were then explicitly converted to integer types (0s and 1s) for consistency.
